- Race – the character's race.
- Gender - the character's gender.
- Age - the character's age.
- Category – the general classification category of the character.
- Affiliation - the group or organization to which the character belongs.
- Alias – alternate names or nicknames of the character.
- Abilities - special abilities of the character.
- Breathing Style - a certain breathing style used by the character in battle.
- Fighting Style – the general fighting method that the character uses.
- Demon Art – special techniques used by demons.
- Equipment - weapons or equipment that the character uses.
- Fights – battles that the character participated in.

```python
#to make the work on this DataFrame easier I prefered to cut the spaces in the columns and than i noticed there is even a wrong space in one column.
ch.rename(columns={'Character No.':'ChNum','Character Name':'ChName','Breathing Style ':'BreathingStyle','Fighting Style':'FightingStyle','Demon Art':'DemonArt'}, inplace=True)
```

```python
# lets see it now
ch.columns
```

```python
# lets check whats is that one NULL (NaN) in Breathing Style column when i checked the data frame info
ch[["ChName","BreathingStyle"]]
```

```python
# Ok i can change it into a "-" like the others who doesnt got a Breathing style
# Another value i can fix better its the "Hinokami Kagura" because in Demon Slayer series it means "Sun Breathing" so lets just change it to english

ch["BreathingStyle"].fillna("-", inplace=True)
ch["BreathingStyle"].replace("Water Breathing, Hinokami Kagura", "Water Breathing, Sun Breathing", inplace=True)
```

```python
# no more nulls?
ch.info()
```

```python
# checking for the changes of Hinokami to Sun
ch.BreathingStyle
```

```python
# I dont like the idea of using diffrent signs for "empty value"
# and till now we can see in some columns "-" and in others "Unknown" so lets decide for one specific value
ch.replace("Unknown","None",inplace=True)
ch.replace("-","None",inplace=True)
```

```python
ch.groupby("Race").ngroups
```

```python
# I noticed that there is from some reason 3 kinds of Races altough it shows me TWO kinds; Demon and Demon Slayer
ch.groupby("Race").groups
```

```python
# So i cutted the useless space in the 13index cause it was written 'Demon '
ch["Race"]=ch.Race.str.strip()
```

```python
# I checked if its better now
ch.groupby("Race").ngroups
```

```python
ch.groupby("Race").groups
```

```python
# I would like to kmow how many excatly Humans and Demons there is in this series
ch["Race"].value_counts()
```

```python
# Lets see how many Males and Females there are in every race
ch.groupby("Race")[["Gender"]].value_counts()
```

```python
gender_race_counts = ch.groupby(["Race", "Gender"]).size().unstack()
plt.figure(figsize=(7, 5))
sb.heatmap(gender_race_counts, cmap="coolwarm", annot=True)
plt.title("Gender Distribution by Race")
plt.xlabel("Gender")
plt.ylabel("Race")
plt.show()
```

```python
# Whats the youngest age and who is it?
youngest=ch.Age.min()
print('The youngest age is',youngest)
ch[ch["Age"]==youngest][["ChName","Age"]]
```

```python
# Whats the oldest age and who is it?
oldest=ch.Age.max()
print('The oldest age is',oldest)
ch[ch["Age"]==oldest][["ChName","Age"]]
```

```python
# Is the age of demons overall greater than the age of humans?
ch.groupby("Race")[["Age"]].sum()
```

```python
# whats the ages of all chracters and whats the count of them in every age
ch.Age.value_counts().sort_index().plot(kind='bar',ylabel='Count',color='skyblue',rot=45)
plt.show()
```

```python
# can I assume that the oldest you get the higher your rank is?
ch[ch["Race"]=="Demon"][["Race","Age","Category"]]
```

```python
plt.figure(figsize=(8,5))
agerank=ch[ch["Race"]=="Demon"][["Race","Age","Category"]]
sb.lineplot(data=agerank, x="Category", y="Age")
plt.show()
```

```python
# How many of the humans characters own breathing style
ch[ch["Race"]=="Human"][["Race","BreathingStyle","Gender"]]
```

```python
# Can we assume that the adoption of a breathing style is influenced by gender?

Gstyle=ch[ch["Race"]=="Human"][["Race","BreathingStyle","Gender"]]

Gstyle_filtered = Gstyle[Gstyle["BreathingStyle"] != "-"]

counts = Gstyle_filtered.groupby("Gender")["BreathingStyle"].count()

counts.plot(kind="pie",autopct='%.0f%%')
plt.show()
```

```python
# What kinds of categories the characters are splitted to?
ch.groupby("Category").groups.keys()
```

```python
# With a graph we can get a better view especially on how the enemies side is way greater than the demon slayers
plt.figure(figsize=(8,5))
sb.countplot(data=ch, y="Category",color='lightgreen')
plt.show()
```

```python
# Lets start and see if the using of Katana as a weapon depends if you got Swordsman\Swordswoman's fighting style
katana_count = ch["Equipment"].str.contains("Katana").sum()
print('The amount users of Katana is:',katana_count)
```

```python
swords_count = ch["FightingStyle"].str.contains("Swords").sum()
print(r'The amount holding the swordsman\woman fighting style is:',swords_count)
```

```python
# With using the shape method i will get the count of the rows with the values katana and swords together 
both_count = ch[(ch["Equipment"].str.contains("Katana")) & 
                (ch["FightingStyle"].str.contains("Swords"))].shape[0]
print(r'The number of rows where they got in the same row the values Katana and Swords(man\woman) is:',both_count)
```

```python
# Now, if the count of the rows with Katana value equal to the count of the rows where Katana appears with
# Swords(man\woman) together that means that holding a Katana depends if you are a Swordsman\woman.
print('Are they equal?',both_count == katana_count)
```

```python
# Lets start and see how many fights every character got into
ch["Fights"]
```

```python
# "VS" is the start of every fight specified in the Fights column.
# I will use this to create a new column that will count for me the amount of "VS"
# for each character in the Fights column
ch['CountFight']=ch['Fights'].str.count('VS')
```

```python
SortedCF = ch.sort_values("CountFight", ascending=False)

plt.figure(figsize=(10, 6))
sb.barplot(y="ChName", x="CountFight", data=SortedCF, palette="Blues_r")

plt.title("Number of Fights per Character", fontsize=14, fontweight="bold")
plt.xlabel("Number of Fights", fontsize=12)
plt.ylabel("Character", fontsize=12)
plt.grid(axis="x", linestyle="--", alpha=0.6)
plt.show()
```
